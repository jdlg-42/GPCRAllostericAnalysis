# Análisis de datos de atención tras analizar con ESM-2

Tras realizar el análisis usando ESM-2 se generan dos archivos CSV, uno para cada receptor, con los que se obtienen datos de interés:

- Número de residuos que superan el percentil 85
- Valor máximo y mínimo
- Valor promedio
- Desviación estándar

## Receptor A2A de adenosina

```python
# Importación del dataset

import pandas as pd

df_a2a = pd.read_csv('accumulated_attention_a2a.csv')
df_a2a
```

```python
# Nº de residuos que superan el percentil 85

print(f"El número de residuos que superan el percentil 95 es de {len(df_a2a)} residuos.")
```

```python
# Ordenar los residuos por valor de atención promedio acumulada

df_a2a.sort_values(by=['AvgAccAtt'], ascending=False)

# Calcular máximo y mínimo

## Obtener las columnas de los valores máximo y mínimo respectivamente
max_row_a2a = df_a2a.loc[df_a2a['AvgAccAtt'].idxmax()]
min_row_a2a = df_a2a.loc[df_a2a['AvgAccAtt'].idxmin()]

## Obtener los valores máximo y mínimo
max_a2a_value = max_row_a2a['AvgAccAtt']
max_residue_a2a = max_row_a2a['Position']

min_a2a_value = min_row_a2a['AvgAccAtt']
min_residue_a2a = min_row_a2a['Position']

print(f"El valor máximo de atención acumulada es de {round(max_a2a_value,4)} para el residuo {max_residue_a2a} y el valor mínimo es de {round(min_a2a_value,4)} para el residuo {min_residue_a2a}.")
```

```python
df_a2a.sort_values(by=['AvgAccAtt'], ascending=False)
```

```python
# Promedio y desviación estandar de atención promedio acumulada

avg_a2a = round(df_a2a['AvgAccAtt'].mean(),4)

print(f"El valor promedio de la atención promedio acumulada es de {avg_a2a}.")

std_a2a = round(df_a2a['AvgAccAtt'].std(),4)

print(f"La desviación estándar de la atención promedio acumulada es de {std_a2a}.")
```
